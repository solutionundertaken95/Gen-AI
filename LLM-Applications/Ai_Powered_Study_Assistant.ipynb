Import module from Gemini and also userdata that will keep our secret keys safe

```python
from google import genai
from google.colab import userdata
```

Use Clientclass for interacting with Gemini and use userdata to insert our secret key securely

---



```python
client = genai.Client(api_key=userdata.get("GEMINI_KEY"))
print(client)
```

Use client object to communicate with Gemini LLM

```python
def study_assistant(user_prompt):
  response = client.models.generate_content(model = 'gemini-2.5-flash',contents=user_prompt)
  return response
```

Call the function

```python
output = study_assistant("Explain Generative AI")
print(output.text)
```
